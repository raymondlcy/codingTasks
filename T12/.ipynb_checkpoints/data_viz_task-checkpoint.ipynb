```python
# Import libraries
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sea

# Inline indicates graphs should be presented as a cell output
%matplotlib inline
```

```python
# Load data
car_data = pd.read_csv('Cars93.csv', index_col = 0, keep_default_na=False) 
```

### 1. Boxplot

```python
# Boxplot for revs per mile

# Get Audi,Hyundai,Suzuki,Toyota data of Rev per mile
audi_rev = car_data[car_data.Manufacturer == 'Audi'].loc[:,"Rev.per.mile"]
#print(audi_rev)
hyundai_rev = car_data[car_data.Manufacturer == 'Hyundai'].loc[:,"Rev.per.mile"]
#print(hyundai_rev)
suzuki_rev = car_data[car_data.Manufacturer == 'Suzuki'].loc[:,"Rev.per.mile"]
#print(suzuki_rev)
toyota_rev = car_data[car_data.Manufacturer == 'Toyota'].loc[:,"Rev.per.mile"]
#print(toyota_rev)

graph_content = [audi_rev,hyundai_rev,suzuki_rev,toyota_rev]
fig, axe = plt.subplots()
axe.boxplot(graph_content)
# Set x label for each Manufacturer
plt.xticks([1, 2, 3, 4], ["Audi", "Hyundai", "Suzuki", "Toyota"])
plt.xlabel("Manufacturer")
plt.ylabel("Rev per mile")
plt.show()
```

**Boxplot question:**  
Which of these manufacturers has the car with the highest revs per mile?


**Boxplot answer:**
Toyota has the car with the highest revs per mile.



### 2. Histogram

```python
# Histogram of MPG
mpg_city = car_data.loc[:,"MPG.city"]
mpg_highway = car_data.loc[:,"MPG.highway"]
 
# Creating a stacked histogram
plt.hist([mpg_city, mpg_highway], stacked=True, color=['cyan', 'Purple'], edgecolor='black')
 
# Adding labels and title
plt.xlabel('MPG')
plt.ylabel('Frequency')
plt.title('MPG City vs MPG Highway')
 
# Adding legend
plt.legend(['MPG.city', 'MPG.highway'])
 
# Display the plot
plt.show()
```

**Histogram question:**  
Is it generally more fuel efficient to drive in the city or on the highway?

**Histogram answer:**
In general, it is more fuel effcient to drive on the highway. The frequency of higher MPG in purple appear more at the right side. it reflect that more car model reach higher MPG than cyan one, which is driving in the city
