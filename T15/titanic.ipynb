```python
import pandas as pd
import numpy as np
# for Box-Cox Transformation
from scipy import stats

# for min_max scaling
from mlxtend.preprocessing import minmax_scaling

# plotting modules
import seaborn as sns
import missingno

import matplotlib.pyplot as plt

# Read Titanic data
titan_df = pd.read_csv("titanic.csv")

# Check total number of record
print(titan_df.shape)
# Find missing data points per column
missing_values_count = titan_df.isnull().sum()
# look at the # of missing points in all columns
print(missing_values_count[0:12])

# Plot graphic of missing values
missingno.matrix(titan_df, figsize = (30,10))
```

### Missing record analysis 

For finding missing value, 177 in 891 are missing Age, so that the age analysis will be less accurate but dropping those record in age related analysis should be still relavant. It is 19% ratio of missing values in total records.

For Cabin, it should show the exact location of the passenger which is not so meaningful so we can ignore and drop it.

For Embarked, the missing data is just two entries, so we can ignore it.

Below we drop the variable cabin and Ticket. Name from the data frame

```python
titan_df.drop('Cabin',axis=1, inplace=True)
titan_df.drop('Ticket',axis=1, inplace=True)
titan_df.drop('Name',axis=1, inplace=True)
titan_df.head()
```

```python

# Retrieve the dataset of surviror and dead
df_survived = titan_df[titan_df.Survived == 1]      # for Survivor
df_dead = titan_df[titan_df.Survived == 0]       # for dead
print(f"Survived Total:{df_survived.shape}")
missing_values_count = df_survived.isnull().sum()
# look at the # of missing points in all columns
print(f"Survived Missing Column:{missing_values_count[0:12]}")

print(f"Dead Total:{df_dead.shape}")
missing_values_count = df_dead.isnull().sum()
print(f"Dead Missing Column:{missing_values_count[0:12]}")

```

And above has already show the overall important index for the records

## General Assumption of defining advantage in group
If the survival rate is more than 38%, that group has advantage in surviving in Titanic incident


```python
# General vs Survival vs Dead by Sex distribution

fig = plt.figure(figsize=(20,10))

ax1 = sns.countplot(x='Sex',  data=titan_df, color='black', alpha=0.85)
for p in ax1.patches:
    ax1.annotate(f'\n{p.get_height()}', (p.get_x()+0.2, p.get_height()+10), ha='center', va='top', color='white', size=10)
ax2= sns.countplot(x='Sex',  data=df_dead, color='red', alpha=0.85);
for p in ax2.patches:
    ax2.annotate(f'\n{p.get_height()}', (p.get_x()+0.2, p.get_height()+10), ha='center', va='top', color='white', size=10)
ax3 = sns.countplot(x='Sex', data=df_survived, color='blue',alpha=0.85);
for p in ax3.patches:
    ax3.annotate(f'\n{p.get_height()}', (p.get_x()+0.2, p.get_height()+10), ha='center', va='top', color='white', size=10)

plt.title('Overall vs Survived and Dead Passenger Grouped by Sex')
plt.legend(['Overall','Dead', 'Survived'])
```


 ## Compare the ratio in survived person by Sex

 From the graphs above, 

 <ul>
 <li>Black is overall passenger</li>
 <li>Blue is survived passenger</li>
 <li>Red is dead passenger</li>
 <li>Purple is the overlapping region of dead and survived. </li>
 </ul>

 </b>

## Result and Conclusion on Survial by Sex

 Male:
 Survival rate is around 19% (109/577), less than the average 38%

 Female:
 Survival rate is around 74% (233/314), more than the average 38%

So it reflect there is trend that female should have higher priority in boarding the lifeboat in order that it is obvious that female got a very high survival rate, and on the reverse, male got far lower survival rate.
