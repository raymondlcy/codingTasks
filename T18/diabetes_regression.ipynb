### Initial Analysis
From the initial analysis, Sex value are only binary and the value for progression are too diverse, and also the prediction between 1 and 2 value are very close. So it is not selected as independent variables.

And from the heatmap, S1 and S2 has very high correlation, so S2 will be dropped to reduce complexity in the calculation

```python
# As SEX is either 1 or 0, which we can't use, so we drop it from the dataset
#df_train = df_diabetes
df_train = df_diabetes.drop('SEX', axis=1)

# drop unwanted attribute
#df_train = df_train.drop('AGE', axis=1)
#df_train = df_train.drop('S1', axis=1)
df_train = df_train.drop('S2', axis=1)
#df_train = df_train.drop('S3', axis=1)
#df_train = df_train.drop('S4', axis=1)
#df_train = df_train.drop('S6', axis=1)

sns.pairplot(data=df_train)
```

### Insight on the feature values
Some of the values do not follow Gaussian model (S4), so that using MinMaxScaler for normalization of the data will be the best choice

```python
# Feature matrix and target vector
X_feature = df_train.drop('PROGRESSION', axis=1)  # We assume 'PROGRESSION' is the column we want to predict
y_target = df_train.iloc[:,8].values  # columns 8 is PROGRESSION
y_target = y_target.reshape(-1, 1) # create 2D matrix

#print("Before")
#print(X_feature)
#print(y_target)

from sklearn.model_selection import train_test_split

# Split the dataset into training and testing sets by 80%/20%
# Use Shuffle = true to make the distribution evenly between test data set and train data set
# Set Same random seed to stablize the same result
rseed = 77
X_train, X_test, y_train, y_test = train_test_split(X_feature, y_target, test_size=0.2, shuffle=True, random_state=rseed)


#print("After")
#print(X_test)
#print(y_test)
```

```python
# Fit the MinMaxScaler on train data
sc = MinMaxScaler()
#sc = StandardScaler()
sc.fit(X_train)

# Apply the scaler on train and test data
X_train = sc.transform(X_train)
X_test = sc.transform(X_test)
```

```python
# Fit a model
lm = LinearRegression()
model = lm.fit(X_train, y_train)
predictions = lm.predict(X_test)

# Create line coordinates
X_line = np.empty(X_test.shape)
for i in range(X_test.shape[1]):
    X_line[:, i] = np.linspace(np.min(X_test[:,i]), np.max(X_test[:,i]), num=X_test.shape[0])
y_line = lm.predict(X_line)

# Display the Intercept and Coeffients
print('Intercept: \n', lm.intercept_)
print('Coefficients: \n', lm.coef_)
```

```python
## Plot the errors
fig, ax = plt.subplots(1, X_test.shape[1], sharey=True, sharex=True, figsize=(8,2))
fig.suptitle("Y vs X on Each Dimension")
ax[0].set_ylabel(df_train.columns[8])

# Get values for the error bar
error_bar_values = np.abs((y_test-predictions)[:,0])

# Plot data, predicted values, and error bars
for i in range(X_test.shape[1]):
    ax[i].set_xlabel(X_feature.columns[i])
    ax[i].errorbar(X_test[:, i], y_test[:, 0], yerr=error_bar_values, fmt='.k', ecolor='red', label='True')
    ax[i].scatter(X_test[:,i], predictions[:,0], c='b', marker='.', label='Predicted')
    ax[i].legend(loc='best', fontsize='x-small')
```

From the error and prediction graph above, we can observe the error are frequet in all x values.

```python
# R squared score

# Display R squared score
print ("R2 Score:", round (model.score(X_test, y_test), 4))
```

## Conclusion
The R2 score is almost 0.5, which mean the test model is not perfect but still can predict the trend for progression with some accuracy. The model should be adjusted and reconsidered about the test and training ratio.
